## 4. 编写可复用清洗函数

函数要求：

- 不直接修改传入的原始 DataFrame；
- 返回 cleaned_df 和 cleaning_log；
- 日志至少包含处理步骤、处理规则、处理前记录数、处理后记录数、影响记录数；
- 完成重复值处理、缺失值处理、类别标准化和必要的数据类型转换。

```python
NUMERIC_MISSING_COLS = [
    'Tenure', 'CityTier', 'WarehouseToHome', 'HourSpendOnApp',
    'NumberOfDeviceRegistered', 'SatisfactionScore', 'NumberOfAddress',
    'Complain', 'OrderAmountHikeFromlastYear', 'CouponUsed',
    'OrderCount', 'DaySinceLastOrder', 'CashbackAmount'
]
CATEGORY_MAPPINGS = {
    "PreferredLoginDevice": {"Phone": "Mobile Phone", "Mobile": "Mobile Phone"},
    "PreferredPaymentMode": {"COD": "Cash on Delivery", "CC": "Credit Card"}
}

def clean_ecommerce_data(data):
    """
    清洗电商用户行为数据。

    参数：
        data: 原始用户行为 DataFrame

    返回：
        cleaned_df: 清洗后的 DataFrame
        cleaning_log: 处理日志 DataFrame
    """
    # TODO：复制数据，避免覆盖原始数据
    # TODO：创建日志列表 logs
    # TODO：删除完全重复行，并记录日志
    # TODO：对 NUMERIC_MISSING_COLS 使用中位数填补，并记录每列影响数量
    # TODO：对 CATEGORY_MAPPINGS 完成类别标准化，并记录每条映射影响数量
    # TODO：将 Churn 和 Complain 转为整数类型
    # TODO：返回 cleaned_df 与 cleaning_log
   
 # TODO：1. 创建数据副本，不修改原始传入data
    cleaned_df = data.copy()
    logs = []  # TODO：2. 创建日志列表存储每一步记录

    # ---------------------- 步骤1：删除完全重复行 ----------------------
    step_name = "删除完全重复记录"
    rule_desc = "删除所有字段取值完全一致的重复行，仅保留第一条"
    before_cnt = len(cleaned_df)
    dup_mask = cleaned_df.duplicated()
    drop_cnt = dup_mask.sum()
    cleaned_df = cleaned_df[~dup_mask]
    after_cnt = len(cleaned_df)
    logs.append({
        "处理步骤": step_name,
        "处理规则": rule_desc,
        "处理前记录数": before_cnt,
        "处理后记录数": after_cnt,
        "影响记录数": drop_cnt
    })

    # ---------------------- 步骤2：数值字段中位数填充缺失值 ----------------------
    step_name = "数值字段中位数填充缺失"
    rule_desc = f"对字段{NUMERIC_MISSING_COLS}使用同列中位数填补空值"
    before_cnt = len(cleaned_df)
    total_impact = 0
    for col in NUMERIC_MISSING_COLS:
        if col in cleaned_df.columns:
            # 先转为数值
            cleaned_df[col] = pd.to_numeric(cleaned_df[col], errors="coerce")
            missing_num = cleaned_df[col].isna().sum()
            med = cleaned_df[col].median()
            cleaned_df[col].fillna(med, inplace=True)
            total_impact += missing_num
    after_cnt = len(cleaned_df)
    logs.append({
        "处理步骤": step_name,
        "处理规则": rule_desc,
        "处理前记录数": before_cnt,
        "处理后记录数": after_cnt,
        "影响记录数": total_impact
    })

    # ---------------------- 步骤3：类别字段标准化映射 ----------------------
    step_name = "类别同义值标准化"
    rule_desc = "按预设字典统一分类字段缩写与全称，消除同义不同值"
    before_cnt = len(cleaned_df)
    total_impact = 0
    for col, map_dict in CATEGORY_MAPPINGS.items():
        if col in cleaned_df.columns:
            # 统计本次映射会改动的行数
            mask = cleaned_df[col].isin(map_dict.keys())
            change_num = mask.sum()
            total_impact += change_num
            cleaned_df[col] = cleaned_df[col].replace(map_dict)
    after_cnt = len(cleaned_df)
    logs.append({
        "处理步骤": step_name,
        "处理规则": rule_desc,
        "处理前记录数": before_cnt,
        "处理后记录数": after_cnt,
        "影响记录数": total_impact
    })

    # ---------------------- 步骤4：指定字段转为整数类型 ----------------------
    step_name = "数据类型强制转换"
    rule_desc = "将Churn、Complain二分类字段转换为int整数型"
    before_cnt = len(cleaned_df)
    target_int_cols = ["Churn", "Complain"]
    impact_rows = 0
    for c in target_int_cols:
        if c in cleaned_df.columns:
            # 转换前非空数值行数即为受影响行数
            valid = cleaned_df[c].notna().sum()
            impact_rows += valid
            cleaned_df[c] = pd.to_numeric(cleaned_df[c], errors="coerce").astype(int)
    after_cnt = len(cleaned_df)
    logs.append({
        "处理步骤": step_name,
        "处理规则": rule_desc,
        "处理前记录数": before_cnt,
        "处理后记录数": after_cnt,
        "影响记录数": impact_rows
    })   # 日志列表转为DataFrame
    cleaning_log = pd.DataFrame(logs)
    # TODO：返回清洗后数据与日志
    return cleaned_df, cleaning_log
     
```

## 5. 数据转换与候选异常值检查

为便于第五天分析，请新增：

- TenureGroup：用户使用时长分层；
- IsMobileLogin：是否主要使用移动端登录；
- 候选异常值报告：WarehouseToHome、OrderCount、CashbackAmount。

候选异常值只记录，不在本项目中自动删除。

```python
def iqr_outlier_summary(series):
    """输出 IQR 候选异常值摘要。"""
    series = series.dropna()
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    return {
        "Q1": q1,
        "Q3": q3,
        "下限": lower,
        "上限": upper,
        "候选异常值数量": int(((series < lower) | (series > upper)).sum())
    }

# TODO：构建 tenure_bins、tenure_labels，并用 pd.cut 新建 TenureGroup
# TODO：新建 IsMobileLogin，移动端为 1，其他设备为 0
# TODO：生成 outlier_report（每行对应一个待检查字段）
# ===================== 1. 新建 TenureGroup：用户使用时长分层 =====================
# 分层区间与标签（电商通用分层逻辑）
enure_bins = [0, 12, 24, 36, 48, 60, float("inf")]
tenure_labels = ["0-12个月", "13-24个月", "25-36个月", "37-48个月", "49-60个月", "60个月以上"]
cleaned_df["TenureGroup"] = pd.cut(
    cleaned_df["Tenure"],
    bins=tenure_bins,
    labels=tenure_labels,
    right=True
)

# TODO：新建 IsMobileLogin，移动端为 1，其他设备为 0
cleaned_df["IsMobileLogin"] = cleaned_df["PreferredLoginDevice"].apply(
    lambda x: 1 if x == "Mobile Phone" else 0
)

# TODO：生成 outlier_report（每行对应一个待检查字段）
check_target_cols = ["WarehouseToHome", "OrderCount", "CashbackAmount"]
report_list = []
for col_name in check_target_cols:
    if col_name in cleaned_df.columns:
        summary_dict = iqr_outlier_summary(cleaned_df[col_name])
        summary_dict["字段名"] = col_name
        report_list.append(summary_dict)
outlier_report = pd.DataFrame(report_list)
print("==== IQR候选异常值检测报告 ====")
print(outlier_report)
```
